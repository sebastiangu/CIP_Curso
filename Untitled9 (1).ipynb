```python
datos = tabl.dropna()
datos.info()
```

```python
print(datos['ocean_proximity'].unique())
```

```python
datos["ocean_proximity"].value_counts()
```

```python
#Dummies / One-Hot Encoding
dummies_ocean = pd.get_dummies(datos["ocean_proximity"], dtype=int)
datos = datos.join(dummies_ocean)
datos.head()
```

```python
datos = datos.drop(["ocean_proximity"], axis=1)
datos.head()
```

```python
datos.corr()
sb.set(rc={'figure.figsize': (15,8)})
sb.heatmap(datos.corr(), annot=True, cmap="YlGnBu")
datos.corr()["median_house_value"].sort_values(ascending=False)
```

```python

X = datos.drop(["median_house_value"], axis=1)
Y = datos["median_house_value"]

from sklearn.model_selection import train_test_split
X_entrena, X_prueba, Y_entrena, Y_prueba = train_test_split(X, Y, test_size=.2)
```

```python
X_prueba.shape
```

```python
from sklearn.linear_model import LinearRegression

modelo=LinearRegression()
modelo.fit(X_entrena, Y_entrena)
```

```python
prediccion = modelo.predict(X_prueba)
print(prediccion)
```

```python
comparativa = {"Prediccion": prediccion, "Valor Real": Y_prueba}
pd.DataFrame(comparativa)
```

```python
#Aplicamos a nuestro modelo
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_prueba, prediccion)
mse
```

```python
# Evaluar el modelo con validación cruzada (métrica: error cuadratico)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(modelo, X, Y, cv=10, scoring='neg_mean_squared_error')
scores
```

```python
import numpy as np
rmse = np.sqrt(mse)
rmse
```

```python
def mostrar_scores(scores):
  print("Puntajes de MSE en cada fold:", scores)  #error cuadratico
  print("Promedio de MSE:", np.mean(scores))  # hallar la media
  print("Desviación estándar de MSE:", np.std(scores))  #desviacion estandar

mostrar_scores(scores)
```

```python
from sklearn.tree import DecisionTreeRegressor
modelDecisionTree = DecisionTreeRegressor(random_state=42)
modelDecisionTree.fit(X_entrena, Y_entrena)
```

```python
scores = cross_val_score(modelDecisionTree, X, Y, cv=10, scoring='neg_mean_squared_error')
scores
```
